```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
fig=plt.figure()
ax=plt.axes()
```

```python


labels=['G1','G2','G3','G4','G5']
men_means = [20,34,30,35,27]
women_means = [25,32,34,20,25]
x=np.arange(len(labels))
width=0.50
fig, ax = plt.subplots()
rects1 =ax.bar(x-width/2,men_means,width,label='men')
rects2 =ax.bar(x+width/2,women_means,width,label='women')

ax.set_ylabel('scroes')
ax.set_ylabel('groups')
ax.set_title('scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend();
```

```python
fig=plt.figure()
ax=plt.axes()


rects1 =ax.bar(x-width/2,men_means,width,label='men')
rects2 =ax.bar(x+width/2,women_means,width,label='women')
width=0.10
ax.set_ylabel('scroes')
ax.set_ylabel('groups')
ax.set_title('scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1,padding=3)
ax.bar_label(rects2,padding=30)
ax.set_ylim(0,40);
```

```python
fig, ax=plt.subplots()
ax.grid(linestyle='--',color='0.95', axis='x');
ax.set_axisbelow(True)
width=0.50
p1 =ax.bar(labels,men_means,width,label='men')
p2 =ax.bar(labels,women_means,width,bottom=men_means,label='women')
ax.set_ylabel('scroes')
ax.set_xlabel('groups')
ax.set_title('scores by group and gender')
ax.legend()

ax.bar_label(p1, label_type='center',color='y')
ax.bar_label(p2,label_type='center',color='k')
ax.bar_label(p2)
ax.set_ylim(0,100);
```

```python
x=np.random.randn(100)
y=np.random.randn(100)
plt.style.use('seaborn-darkgrid')
z1=np.sqrt(x**2 + y**2)
fig, ax=plt.subplots()
pos=ax.scatter(x,y ,c=z1,cmap='cool',marker ='3')
fig.colorbar(pos);
```

```python
fig, ax=plt.subplots()
pos=ax.scatter(x,y ,c=z1,s=100,alpha = 0.50, cmap='viridis',)
fig.colorbar(pos);
```

```python
spread=np.random.rand(50)*100
center=np.ones(25)*100
filer_high = np.random.rand(10) *100+100
filer_low = np.random.rand(10) * -100
data=np.concatenate((spread,center,filer_high ,filer_low))
data
```

```python
fig ,ax=plt.subplots()
ax.boxplot(data)
ax.set_title('basic box plot')
```
